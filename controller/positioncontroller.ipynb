```python
import serial, time
import math
import numpy as np
import requests


ser = serial.Serial("/dev/serial0", 115200,timeout=0) # mini UART serial device

def read_tfluna_data():
    while True:
        counter = ser.in_waiting # count the number of bytes of the serial port
        if counter > 8:
            bytes_serial = ser.read(9) # read 9 bytes
            ser.reset_input_buffer() # reset buffer

            if bytes_serial[0] == 0x59 and bytes_serial[1] == 0x59: # check first two bytes
                distance = bytes_serial[2] + bytes_serial[3]*256 # distance in next two bytes
                strength = bytes_serial[4] + bytes_serial[5]*256 # signal strength in next two bytes
                temperature = bytes_serial[6] + bytes_serial[7]*256 # temp in next two bytes
                temperature = (temperature/8.0) - 256.0 # temp scaling and offset
                return distance/100.0,strength,temperature
            
def get_distance():
    while True:
        counter = ser.in_waiting
        if counter > 8:
            bytes_serial = ser.read(9)
            ser.reset_input_buffer()
            
            if bytes_serial[0] == 0x59 and bytes_serial[1]==0x59:
                distance = bytes_serial[2] + bytes_serial[3]*256
                return distance/100.0
def move_step(amt):
    req = requests.post("http://192.168.1.42:8000/", data={'direction': "forward" if amt >= 0 else 'backward', "step_count": amt})
def set_angle(angle):
    req = requests.post("https://192.168.1.42:8000/servo_B", data={'angle' : angle})
    




def move_to_pos(pos):
    #pid control?
    distance = get_distance()
    diff = pos - distance
    while diff > 0.1:
        distance = get_distance()
        diff = pos - distance
        move_step(20) #calculate steps here
        time.sleep(1)
      
        
if ser.isOpen() == False:
    ser.open() # open serial port if not open


#main loop?
#while True:

segments = 10
#total distance = 5.4 m
def steps_to_m(steps):
    return steps * 10

#main loop
longitude = 0
latitude = 0

degrees_per_lon_segment  = 30
def cycle_lon():
    for i in range(math.floor(360/degrees_per_lon_segment)):
        longitude = i * degrees_per_lon_segment - 180
        print(longitude)
       # set_angle(longitude)
        time.sleep(1)

degrees_per_lat_segment = 30
def cycle_lat():
    for i in range(math.floor(360/degrees_per_lat_segment)):
        latitude = i * degrees_per_lat_segment - 180
        print(latitude)
        #move_to_pos(latitude) #need convert to steps or something
        time.sleep(1)

#we can wait a bit before starting each longitude segment to cover everything i think

ser.close() # close serial port
```
